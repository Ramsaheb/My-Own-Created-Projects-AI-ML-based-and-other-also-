```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader('https://nvidia.wd5.myworkdayjobs.com/en-US/NVIDIAExternalCareerSite/job/DFX-Software-QA-Test-Dev-Engineer_JR1998610?locationHierarchy1=2fcb99c455831013ea52b82135ba3266')

page_data = loader.load().pop().page_content
print(page_data)
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get('https://nvidia.wd5.myworkdayjobs.com/en-US/NVIDIAExternalCareerSite/job/DFX-Software-QA-Test-Dev-Engineer_JR1998610?locationHierarchy1=2fcb99c455831013ea52b82135ba3266')
page_data = driver.page_source
print(page_data)
driver.quit()
```

```python
from langchain_core.prompts import PromptTemplate
from IPython.display import Markdown


promt_extract = PromptTemplate.from_template('''

You are a job description extractor. Extract the following information from the job description:
1. Job Title
2. Company Name
3. Location
4. Job Description
5. Required Skills
6. Preferred Skills
7. Responsibilities
8. Qualifications
9. Salary Range
10. Benefits
Extract the information from the following job description:
{job_description}
''')


chain_extract = promt_extract | llm
res = chain_extract.invoke({'job_description': page_data})

Markdown(res.content)
```

```python
import pandas as pd

df = pd.read_csv('my_portfolio.csv')
df
```

```python
import chromadb
import uuid

client = chromadb.PersistentClient('vectorstore')
collection = client.get_or_create_collection(name='portfolio')

if not collection.count():
    for _, row in df.iterrows():
        collection.add(documents=[row['Techstack']],
                       metadatas=[{'link': row['Links']}],
                       ids=[str(uuid.uuid4())])
```

```python
prompt_email = PromptTemplate.from_template(
    '''You are a cold email generator. Generate a cold email for the following portfolio:

    {portfolio}

    '''
)
chain_email = prompt_email | llm
res = chain_email.invoke({'portfolio': df.to_dict(orient='records')})
Markdown(res.content)
```
